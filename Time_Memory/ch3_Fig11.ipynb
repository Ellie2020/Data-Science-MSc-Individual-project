```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot
from matplotlib.pyplot import *
import math
```

```python
MH600B150LSH=pd.read_csv('TIMEandMEM0911.csv')
MH600B150LSH
```

```python
MH600LSH1=MH600B150LSH[MH600B150LSH['LSH'].eq(1)]
MH600LSH1
```

```python
# MH600 B150 LSH1
x=4
import statistics

MH600LSH1B150MeanTime=round(MH600LSH1.TIME_1.mean(), 2)
MH600LSH1B150MeanMEM=round(MH600LSH1.Gigabyte.mean(), 2)
MH600LSH1B150TimeError=round(statistics.stdev(MH600LSH1.TIME_1)/math.sqrt(x),3)
MH600LSH1B150MemError=round(statistics.stdev(MH600LSH1.Gigabyte)/math.sqrt(x),3)


LSH1MH600B150={'MH': 600, 'B':150, 'LSH':1, 'time_mean':MH600LSH1B150MeanTime, 'Gigabyte':MH600LSH1B150MeanMEM, 'Time_1SE':MH600LSH1B150TimeError, 'Mem_1SE':MH600LSH1B150MemError}
```

```python
LSH1MH600B150_df=pd.DataFrame([LSH1MH600B150])
LSH1MH600B150_df
```

```python
MH600LSH3=MH600B150LSH[MH600B150LSH['LSH'].eq(3)]
MH600LSH3
```

```python
# MH600 B150 LSH1
x=4
import statistics

MH600LSH3B150MeanTime=round(MH600LSH3.TIME_1.mean(), 2)
MH600LSH3B150MeanMEM=round(MH600LSH3.Gigabyte.mean(), 2)
MH600LSH3B150TimeError=round(statistics.stdev(MH600LSH3.TIME_1)/math.sqrt(x),3)
MH600LSH3B150MemError=round(statistics.stdev(MH600LSH3.Gigabyte)/math.sqrt(x),3)


LSH3MH600B150={'MH': 600, 'B':150, 'LSH':3, 'time_mean':MH600LSH3B150MeanTime, 'Gigabyte':MH600LSH3B150MeanMEM, 'Time_1SE':MH600LSH3B150TimeError, 'Mem_1SE':MH600LSH3B150MemError}
```

```python
LSH3MH600B150_df=pd.DataFrame([LSH3MH600B150])
```

```python
LSH3MH600B150_df
```

```python
# MH600 B150 LSH1
x=4
import statistics

MH600LSH5B150MeanTime=round(MH600LSH5.TIME_1.mean(), 2)
MH600LSH5B150MeanMEM=round(MH600LSH5.Gigabyte.mean(), 2)
MH600LSH5B150TimeError=round(statistics.stdev(MH600LSH5.TIME_1)/math.sqrt(x),3)
MH600LSH5B150MemError=round(statistics.stdev(MH600LSH5.Gigabyte)/math.sqrt(x),3)


LSH5MH600B150={'MH': 600, 'B':150, 'LSH':3, 'time_mean':MH600LSH5B150MeanTime, 'Gigabyte':MH600LSH5B150MeanMEM, 'Time_1SE':MH600LSH5B150TimeError, 'Mem_1SE':MH600LSH5B150MemError}
```

```python
MH600B150_MEMTIME=pd.concat([LSH1MH600B150_df,LSH3MH600B150_df,LSH5MH600B150_df])
MH600B150_MEMTIME
```

```python
MH600B150_MEMTIME

import matplotlib
import matplotlib.pyplot as plt
from matplotlib.transforms import Affine2D
a=MH600B150_MEMTIME
fig, ax = plt.subplots(figsize=(8,5))

# get handles
handles, labels = ax.get_legend_handles_labels()

x=['1','3','5']

trans1 = Affine2D().translate(-0.005, 0.0) + ax.transData
trans2 = Affine2D().translate(+0.005, 0.0) + ax.transData

y= [a.time_mean.values[0], a.time_mean.values[1], a.time_mean.values[2]]
e = [a.Time_1SE.values[0], a.Time_1SE.values[1], a.Time_1SE.values[2]]

y2 = [a.Gigabyte.values[0], a.Gigabyte.values[1], a.Gigabyte.values[2]]
e2 = [a.Mem_1SE.values[0],a.Mem_1SE.values[1],a.Mem_1SE.values[2]]

er = ax.errorbar(x, y, yerr=e, marker="o", linestyle='dotted', transform=trans1, label='Time')

ax2 = ax.twinx()  

er2 = ax2.errorbar(x, y2, yerr=e2, marker="o", linestyle="dotted", color='r', label='Memory')

ax.set_ylabel('seconds',fontsize=16)
ax2.set_ylabel('gigabyte',fontsize=16)
ax.set_xlabel('LSH',fontsize=14)
ax.legend(bbox_to_anchor=(1.3, 0.9), fontsize=14)
ax2.legend(bbox_to_anchor=(1.3, 1.05), fontsize=14)
ax.tick_params(labelcolor='black', labelsize=14, width=3)
plt.show()
```
